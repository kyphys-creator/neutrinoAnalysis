```python
from neutrino_analysis_band import NeutrinoAnalysis
a = NeutrinoAnalysis(background_scenario='b', intervals='180',
                     GeV=0.6e16, solver='osqp')
a.optimize(a.data_vector)   # 理論曲線とエネルギー軸 (eb) のため

a.plot_band_comparison(
    {'a':    'scenario_bkg_a/bands/band_*.json',
     'flat': 'scenario_bkg_flat/bands/band_*.json',
     'b':    'scenario_bkg_b/bands/band_*.json',},
    level=0.954,            # 2σ
    ylim=(1e11, 1e14),
    save=True, fname='band_comparison_flat_vs_b_2sigma.pdf',
    logy=True
)
```
