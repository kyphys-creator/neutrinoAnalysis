```python
from neutrino_analysis_band import NeutrinoAnalysis
```

```python
# Run optimize() on each scenario first
a_flat = NeutrinoAnalysis(background_scenario='flat', intervals='180', GeV=0.32e16, solver='osqp', T=300)
a_flat.optimize(a_flat.data_vector)
a_a = NeutrinoAnalysis(background_scenario='a', intervals='180', GeV=0.36e16, solver='osqp', T=300)
a_a.optimize(a_a.data_vector)
a_b = NeutrinoAnalysis(background_scenario='b', intervals='180', GeV=0.6e16, solver='osqp', T=300)
a_b.optimize(a_b.data_vector)

a_flat.plot_band_comparison(
    {'Flat Bkg':    'T300/scenario_bkg_flat/bands/band_*.json',},
    level=0.678,
    optimized={'b': a_b},   # Overlay each scenario's optimize result
    ylim=(0, 2e12), save=True,
    fname='band_comparison_T300_1sigma_edge.pdf',
    logy = False
)
```
