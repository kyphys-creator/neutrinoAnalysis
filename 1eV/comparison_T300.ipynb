```python
from neutrino_analysis_band import NeutrinoAnalysis
```

```python
# Run optimize() on each scenario first
a_flat = NeutrinoAnalysis(background_scenario='flat', intervals='180', GeV=0.32e16, solver='scipy', T=300)
a_flat.optimize(a_flat.data_vector)
```

```python
a_flat.plot_band_comparison(
    {r'1$\sigma$ CL: Flat Bkg':    'T300/scenario_bkg_flat/bands/band_*.json',},
    level=0.678,
    optimized={'Flat Bkg': a_flat},   # Overlay each scenario's optimize result
    ylim=(0, 2e12), save=True,
    fname='band_comparison_T300_1sigma.pdf',
    logy = False,
    style = 'both'
)
```

```python
a_flat.plot_band_comparison(
    {r'1$\sigma$ CL: Flat Bkg':    'T300/scenario_bkg_flat/bands/band_*.json',},
    level=0.678,
    optimized={'Flat Bkg': a_flat},   # Overlay each scenario's optimize result
    ylim=(1e11, 4e13), save=True,
    fname='band_comparison_T300_1sigma_edge.pdf',
    logy = True,
    style = 'both'
)
```
