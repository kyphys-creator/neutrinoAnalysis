# 03 · Δχ² scan on a fixed parameter

`scan_fixed_parameter` walks one flux component across a grid, refits with
that component pinned, and runs a Monte-Carlo (Neyman) test at each point
to decide whether the value is inside the 1σ / 90% / 2σ band.

It returns a list of dicts — drop it straight into a DataFrame.

```python
import warnings; warnings.simplefilter('ignore')
from neutrino_analysis_band import NeutrinoAnalysis
import pandas as pd

a = NeutrinoAnalysis(background_scenario='flat', intervals='180',
                     GeV=0.32e16, solver='osqp', T=3)
a.optimize(a.data_vector)   # best fit first; fills a.result

scan = a.scan_fixed_parameter(
    fixed_index=20, scan_range=0.5, num_points=6,
    num_pseudo_data=20, seed=42, n_jobs=1,
)
df = pd.DataFrame(scan)
df[['fixed_value_physical', 'included1sigma', 'included90',
    'included2sigma', 'delta_chi2_cutoff90']]
```

Columns: `fixed_value_physical` is the pinned flux (cm⁻² s⁻¹);
`includedXX` flags whether it falls inside that level; `delta_chi2_cutoffXX`
is the Monte-Carlo cutoff used. This is the manual-grid approach — lesson
04 finds the band edges directly instead.
