## Dependencies

Core: `numpy scipy matplotlib pandas numba`.  
For `solver='osqp'`: `cvxpy osqp` (and optionally `highspy` for the
fastest staircase solver — SciPy's bundled LP is used otherwise).

```python
import warnings
warnings.simplefilter('ignore')   # quiet solver 'inaccurate' notices

from neutrino_analysis_band import NeutrinoAnalysis, load_band
import numpy as np
import cvxpy as cp

print('cvxpy solvers available:', cp.installed_solvers())
# 'SCIPY' guarantees the staircase even without HiGHS
```

```python
# Construct an analysis object. Arguments you will reuse throughout:
#   background_scenario : 'a' / 'b' / 'b2' / 'c' / 'flat' / 'none'
#   intervals           : response-matrix granularity (here '180')
#   GeV                 : natural-unit scale (0.3-0.6e16 is well-conditioned)
#   solver              : 'scipy' (stable) or 'osqp' (fast)
#   T                   : observation time; sets output folder T<T>/
a = NeutrinoAnalysis(background_scenario='flat', intervals='180',
                     GeV=0.32e16, solver='scipy', T=3)
print('bins (m) =', a.m, ' flux params (n) =', a.n)
print('output dir   :', a.scenario_dir)
print('bands dir    :', a.bands_dir)
```
