# 07 · Varying the observation time `T`

`T` is a constructor argument (default 3). It scales the statistics and,
for bookkeeping, prefixes every output path with `T<T>/`, so different
`T` runs never overwrite each other.

```python
import warnings; warnings.simplefilter('ignore')
from neutrino_analysis_band import NeutrinoAnalysis

for T in (3, 30, 300):
    a = NeutrinoAnalysis(background_scenario='flat', intervals='180',
                         GeV=0.32e16, solver='osqp', T=T)
    print('T=%-3d  scenario_dir=%-26s bands_dir=%s'
          % (T, a.scenario_dir, a.bands_dir))
```

Larger `T` gives **narrower** bands (more statistics). Because the bands
shrink, numerical precision matters more: prefer a well-conditioned
`GeV` (~0.3-0.6e16) and a large `n_pseudo_edge`. Lesson 08 covers this.

```python
# Quick illustration: best fit is the same, but compare T's later via
# find_confidence_band with a large n_pseudo_edge. Here just confirm the
# objects build and the output paths differ.
import os
a = NeutrinoAnalysis(background_scenario='flat', intervals='180',
                     GeV=0.32e16, solver='osqp', T=30)
a.optimize(a.data_vector)
a.plot_flux_comparison(save=True)   # writes under T30/
print('exists:', os.path.exists(a.scenario_dir))
```
