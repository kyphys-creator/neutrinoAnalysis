# 06 · Compare scenarios on one axis

`plot_band_comparison` overlays one confidence level from several
scenarios, each in its own colour. Bands are physical units, so they are
comparable across scenarios. `optimized=` adds each scenario's flux
scatter in the matching colour.

This lesson reuses the bands you saved in lesson 05, plus a second
scenario's bands.

```python
import warnings; warnings.simplefilter('ignore')
from neutrino_analysis_band import NeutrinoAnalysis

# Scenario 1: flat (bands from lesson 05). Recompute here so the
# notebook is runnable on its own.
a_flat = NeutrinoAnalysis(background_scenario='flat', intervals='180',
                          GeV=0.32e16, solver='osqp', T=3)
a_flat.optimize(a_flat.data_vector)
for idx in [0, 20, 40, 80, 120]:
    a_flat.find_and_save_band(idx, levels=(0.678, 0.954),
        num_pseudo_data=20, n_pseudo_edge=80, step=1.5, rel_tol=0.08,
        seed=42, verbose=False)

# Scenario 2: background 'a'.
a_a = NeutrinoAnalysis(background_scenario='a', intervals='180',
                       GeV=0.32e16, solver='osqp', T=3)
a_a.optimize(a_a.data_vector)
for idx in [0, 20, 40, 80, 120]:
    a_a.find_and_save_band(idx, levels=(0.678, 0.954),
        num_pseudo_data=20, n_pseudo_edge=80, step=1.5, rel_tol=0.08,
        seed=42, verbose=False)
```

```python
a_flat.plot_band_comparison(
    {'flat': f'{a_flat.bands_dir}/band_*.json',
     'a':    f'{a_a.bands_dir}/band_*.json'},
    level=0.954,
    optimized={'flat': a_flat, 'a': a_a},
    ylim=(0, 3e13), save=True,
)
import matplotlib.pyplot as plt; plt.show()
```

Labels in `groups` and `optimized` should match so the band and its
scatter share a colour. Differences smaller than `rel_tol` are not
physical — see lesson 08.
