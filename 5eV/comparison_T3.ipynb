```python
from neutrino_analysis_band import NeutrinoAnalysis
```

```python
# Run optimize() on each scenario first
a_flat = NeutrinoAnalysis(background_scenario='flat', intervals='180', GeV=0.32e16, solver='scipy')
a_flat.optimize(a_flat.data_vector)
a_a = NeutrinoAnalysis(background_scenario='a', intervals='180', GeV=0.36e16, solver='scipy')
a_a.optimize(a_a.data_vector)
a_b = NeutrinoAnalysis(background_scenario='b', intervals='180', GeV=0.6e16, solver='scipy')
a_b.optimize(a_b.data_vector)
```

```python
a_b.plot_band_comparison(
    {r'1$\sigma$ CL: Exp Bkg':    'T3/scenario_bkg_b/bands/band_*.json',
     r'1$\sigma$ CL: Flat Bkg': 'T3/scenario_bkg_flat/bands/band_*.json',
     r'1$\sigma$ CL: No Bkg': 'T3/scenario_bkg_a/bands/band_*.json',},
    level=0.678,
    optimized={'Exp Bkg': a_b},   # Overlay each scenario's optimize result
    ylim=(0, 2e12), save=True,
    fname='band_comparison_T3_1sigma.pdf',
    logy = False,
    style = 'both'
)
```

```python
a_b.plot_band_comparison(
    {r'1$\sigma$ CL: Exp Bkg':    'T3/scenario_bkg_b/bands/band_*.json',
     r'1$\sigma$ CL: Flat Bkg': 'T3/scenario_bkg_flat/bands/band_*.json',
     r'1$\sigma$ CL: No Bkg': 'T3/scenario_bkg_a/bands/band_*.json',},
    level=0.678,
    optimized={'Exp Bkg': a_b},   # Overlay each scenario's optimize result
    ylim=(1e11, 3e13), save=True,
    fname='band_comparison_T3_1sigma_edge.pdf',
    logy = True,
    style = 'both'
)
```

```python
a_b.plot_band_comparison(
    {r'2$\sigma$ CL: Exp Bkg':    'T3/scenario_bkg_b/bands/band_*.json',
     r'2$\sigma$ CL: Flat Bkg': 'T3/scenario_bkg_flat/bands/band_*.json',
     r'2$\sigma$ CL: No Bkg': 'T3/scenario_bkg_a/bands/band_*.json',},
    level=0.954,
    optimized={'Exp Bkg': a_b},   # Overlay each scenario's optimize result
    ylim=(0, 2e12), save=True,
    fname='band_comparison_T3_2sigma.pdf',
    style = 'both'
)
```

```python
a_b.plot_band_comparison(
    {r'2$\sigma$ CL: Exp Bkg':    'T3/scenario_bkg_b/bands/band_*.json',
     r'2$\sigma$ CL: Flat Bkg': 'T3/scenario_bkg_flat/bands/band_*.json',
     r'2$\sigma$ CL: No Bkg': 'T3/scenario_bkg_a/bands/band_*.json',},
    level=0.954,
    optimized={'Exp Bkg': a_b},   # Overlay each scenario's optimize result
    ylim=(1e11, 3e13), save=True,
    fname='band_comparison_T3_2sigma_edge.pdf',
    logy = True,
    style = 'both'
)
```
