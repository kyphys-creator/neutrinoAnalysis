### Regularized unfolding vs 5 eV Flat Bkg reconstruction

Same layout as `plot_band_comparison` (Flat Bkg 2$\sigma$ band + degeneracy band + best-fit + true flux), with the regularized-unfolding band overlaid in red, reproducing `FinalFigures/cgb_reconstruction_5eV_flat.pdf`.

```python
# Regularized unfolding vs 5eV Flat Bkg reconstruction
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate

norm = 1e12

# base layout from plot_band_comparison: Flat Bkg 2-sigma band + degeneracy
# band + best-fit points + true flux (save=False so we can overlay below)
a_flat.plot_band_comparison(
    {r'2$\sigma$ CL: Flat Bkg':
        'T3/scenario_bkg_flat/bands/band_bkgflat_idx[0-9][0-9][0-9].json'},
    level=0.954,
    optimized={'Flat Bkg': a_flat},
    degeneracy_files='T3/scenario_bkg_flat/degeneracy/degeneracy_bkgflat_idx*.json',
    ylim=(0, 2.5e12), style='fill', save=False,
)

# --- overlay the regularized-unfolding band (scenario 2) ---
upperBand_scenario2 = pd.read_csv('FinalFigures/upperBand_scenario2.csv')
lowerBand_scenario2 = pd.read_csv('FinalFigures/lowerBand_scenario2.csv')
deff = 1 / np.sqrt(1 / (72e2)**2 + 1 / (102e2)**2)
const = (205.3 / 2.65e22) * (4 * np.pi * deff**2)

def upperBand(E):
    return np.interp(E, np.array(upperBand_scenario2['MeV']),
                     np.array(upperBand_scenario2['fissionMeV']))

def lowerBand(E):
    return np.interp(E, np.array(lowerBand_scenario2['MeV']),
                     np.array(lowerBand_scenario2['fissionMeV']))

MeVlist = np.append(np.array(upperBand_scenario2['MeV']), np.array([2]))
upperBandlist = upperBand(MeVlist).astype(float)
lowerBandlist = lowerBand(MeVlist).astype(float)
for i in range(len(MeVlist)):
    upperBandlist[i] = integrate.trapezoid(upperBandlist[i:], MeVlist[i:]) / const
    lowerBandlist[i] = integrate.trapezoid(lowerBandlist[i:], MeVlist[i:]) / const

plt.fill_between(MeVlist, lowerBandlist / norm, upperBandlist / norm,
                 color='#d62728', alpha=0.12, zorder=2)
plt.plot(MeVlist, upperBandlist / norm, color='#d62728', lw=1.6, zorder=3,
         label='Reg. unfolding')
plt.plot(MeVlist, lowerBandlist / norm, color='#d62728', lw=1.6, zorder=3)
plt.legend(loc='upper right', fontsize=15, frameon=False)
plt.savefig('comparison_unfolding_vs_flat_5eV.pdf')
plt.show()
```
